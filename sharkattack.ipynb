```python
df['Activity'] = df['Activity'].astype(str).str.lower()
df['Activity'] = df['Activity'].str.strip()
```

```python
def activity_clean(row):
    surf = re.findall('[s][u][r][f]\w*', row['Activity'])
    swimming = re.findall('[s][w][i][m][m][i][n][g]\w*', row['Activity'])
    fishing = re.findall('[f][i][s][h][i][n][g]\w*', row['Activity'])
    spearfishing = re.findall('[s][p][e][a][r][f][i][s][h][i][n][g]\w*', row['Activity'])
    bathing = re.findall('[b][a][t][h][i][n][g]\w*', row['Activity'])
    wading = re.findall('[w][a][d][i][n][g]\w*', row['Activity'])
    diving = re.findall('[d][i][v][i][n][g]\w*', row['Activity'])
    standing = re.findall('[s][t][a][n][d][i][n][g]\w*', row['Activity'])
    if len(surf) != 0:
        return "SURFING"
    elif len(swimming) != 0:
        return "SWIMMING" 
    elif len(fishing) != 0:
        return "FISHING"  
    elif len(spearfishing) != 0:
        return "SPEARFISHING" 
    elif len(bathing) != 0:
        return "BATHING" 
    elif len(wading) != 0:
        return "WADING"  
    elif len(diving) != 0:
        return "DIVING" 
    elif len(standing) != 0:
        return "STANDING"
    else:
        return "Others"
```

```python
df['Activity'] = df.apply(activity_clean, axis = 1)
```

```python
df['Date'] = df['Date'].astype(str).str.lower()
df['Date'] = df['Date'].str.strip()
```

```python
def year(x):
    res = re.search(r'([0-9]{4})', str(x))
    if res is None:
        return None
    else:
        return int(res.group())
```

```python
def month(x):
    res = re.search(r'(?!-)(([A-Za-z]{3}))(?=-)', str(x))
    if res is None:
        return None
    else:
        return str(res.group())
```

```python
def day(x):
    res = re.search(r'(^|\b)(([0-9]{2}))(?=-)',str(x))
    if res is None:
        return None
    else:
        return str(res.group())
```

```python
df['Day'] = df.Date.apply(day)
```

```python
df['Month'] = df.Date.apply(month)
```

```python
df['Year'] = df.Date.apply(year)
```

```python
def type_clean(row):
    provoked = re.findall('[p][r][o][v][o][k][e][d]\w*', row['Type'])
    boat = re.findall('[b][o][a][t]\w*', row['Type'])
    unprovoked = re.findall('[u][n][p][r][o][v][o][k][e][d]\w*', row['Type'])
    if len(provoked) != 0:
        return "PROVOKED"
    elif len(boat) != 0:
        return "BOAT"
    elif len(unprovoked) != 0:
        return "UNPROVOKED"
    else:
        return "INVALID"
```

```python
df['Type'] = df.apply(type_clean, axis = 1)
```

```python
def species_clean(row):
    whiteshark = re.findall('[w][h][i][t][e][ ][s][h][a][r][k]\w*', row['Species '])
    tigershark = re.findall('[t][i][g][e][r][ ][s][h][a][r][k]\w*', row['Species '])
    bullshark = re.findall('[b][u][l][l][ ][s][h][a][r][k]\w*', row['Species '])
    wobbegongshark = re.findall('[w][o][b][b][e][g][o][n][g][ ][s][h][a][r][k]\w*', row['Species '])
    blacktipshark = re.findall('[b][l][a][c][k][t][i][p][ ][s][h][a][r][k]\w*', row['Species '])
    if len(whiteshark) != 0:
        return "WHITE SHARK"
    elif len(tigershark) != 0:
        return "TIGER SHARK"
    elif len(bullshark) != 0:
        return "BULL SHARK"
    elif len(wobbegongshark) != 0:
        return "WOBBEGONG SHARK"  
    elif len(blacktipshark) != 0:
        return "BLACKTIP SHARK"
    else:
        return "INVALID"
```

```python
df['Species '] = df.apply(species_clean, axis = 1)
```
